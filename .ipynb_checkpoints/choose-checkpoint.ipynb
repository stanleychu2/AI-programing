```python
# return 使用者所需的那一列 注意行數的不同
# head 和 tail 預設都是印出前後的 5 筆資料
def numberRow(row) :
    if(row > df.shape[0] or row <=0) :
        print("抱歉沒有這一列喔！")
    else : 
        front = df.head(n = row)
        answer = front.tail(n = 1)
        return answer
numberRow(10)
```

```python
# 將 DataFrame 轉換為 array
shop = df["店名"].values
price = df["平均價錢(元)"].values
waitTime = df["等候時間(分鐘)"].values
service = df["服務態度(1~5)"].values
foodQuality = df["食物品質(1~5)"].values
location = df["地區(北中南)"].values
meal = df["早餐午餐晚餐其他"].values
typeOfMeal = df["中餐西餐"].values

# 提供給使用者的選擇
scoreChoice = [1,2,3,4,5]
timeChoice = [5,10,20,30]
priceChoice = [50,100,200,300,500]
mealChoice = ["早餐","午餐","晚餐","其他"]
typeOfMealChoice = ["中餐","西餐"]
locationChoice = ["北","中","南","東","全台"]

# 利用使用者的選擇從資料中挑選出合適的餐廳
def displayShop(三餐,中西餐,地區,價錢,等候時間,服務態度,食物品質):
    size = 0
    result = pd.DataFrame()
    for i in range(df.shape[0]):
        if(三餐 in meal[i] and 中西餐 in typeOfMeal[i] and 地區 in location[i]
           and 價錢 >= price[i] and 等候時間 >= waitTime[i] and 服務態度 <= service[i]
           and 食物品質 <= foodQuality[i]):
            result = result.append(numberRow(i+1))
            size+=1
    if(size==0):
        print("非常抱歉沒有符合您的選項的餐廳")
    else:
        print("\n有 " + str(size) + " 家餐廳符合你的要求ㄛ")
        # 另一種對字串的處理方式
        # print("\n有 {} 家餐廳符合你的要求ㄛ".format(size))      
        return result

# 處理互動表單的部分
interact(displayShop,三餐=mealChoice,中西餐=typeOfMealChoice,地區=locationChoice,價錢=priceChoice,等候時間=timeChoice,服務態度=scoreChoice,食物品質=scoreChoice)
```

# 練習題目
### 先將早午、午餐、晚餐從資料中分離出來
### 利用亂數將早午晚餐選出來
### 三餐中不能有重複的餐廳，如有重複請再亂數抽一次
### 假如有其中的一餐超過 X 元便不能吃點心，如果沒有便能吃點心
### np.random.randint(下限,上限) 亂數的使用方法 --> 不包含上限

```python
breakfast = []
lunch = []
dinner = []
snack = []
expensive = []
for i in range(meal.shape[0]):
    if("早餐" in meal[i]):
        breakfast.append(shop[i])
    if("午餐" in meal[i]):
        lunch.append(shop[i])
    if("晚餐" in meal[i]):
        dinner.append(shop[i])
    if("其他" in meal[i]):
        snack.append(shop[i])
    if(price[i] > 150):
        expensive.append(shop[i])
def selectWhatToEat():
    finalChoose = []
    finalChoose.append(breakfast[np.random.randint(0,len(breakfast))])
    repeat(finalChoose, lunch)
    repeat(finalChoose, dinner)
    if(not exceed(finalChoose)):
        repeat(finalChoose, snack)
    return finalChoose
```

```python
selectWhatToEat() # 早餐 午餐 晚餐 點心
```

# 隨機早午晚餐步驟
### 1. 先從早餐的 list 中 random 一個餐廳出來，將挑選出來的答案放進最後結果的 list 裡
### 2. 再來 random 中餐，但這次要多一個步驟，要去檢查是否和早餐相同，如果相同便在random一次，還是相同就在做一次，直到不同為止
### 3. random 晚餐，步驟同上
### 4. 現在我們擁有了一個早午晚餐的 list，接著我們要去看是否有其中一餐超過 150 元，如果有超過的餐廳就直接印出答案，沒有超過的餐廳便代表可以吃點心，需要 random一個點心並加到 list 裡，這裡也要注意的是點心也有可能重複，所以也要做檢查的動作

```python
# 幫你檢查是否有重複的餐廳，有重複便拿掉重複的部分再重來一次，finalChoose 是最後放早午晚餐的那個地方，meal 代表現在處理的是哪一餐
def repeat(finalChoose, meal):
    while(True):
        finalChoose.append(meal[np.random.randint(0,len(meal))])
        if(finalChoose[len(finalChoose)-1] not in finalChoose[:len(finalChoose)-1]):
            break
        finalChoose = finalChoose[:len(finalChoose)-1]
# 檢查有沒有超過 150 元
def exceed(finalChoose):
    for i in finalChoose:
            if i in expensive:
                return True
    return False
```
