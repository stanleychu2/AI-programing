```python
df = pd.read_excel("課程店家.xlsx")
```

```python
# return 使用者所需的那一列 注意行數的不同
# head 和 tail 預設都是印出前後的 5 筆資料
def numberRow(row) :
    if(row > df.shape[0] or row <=0) :
        print("抱歉沒有這一列喔！")
    else : 
        front = df.head(n = row)
        answer = front.tail(n = 1)
        return answer
numberRow(10)
```

```python
# 將 DataFrame 轉換為 array
shop = df["店名"].values
price = df["平均價錢(元)"].values
waitTime = df["等候時間(分鐘)"].values
service = df["服務態度(1~5)"].values
foodQuality = df["食物品質(1~5)"].values
location = df["地區(北中南)"].values
meal = df["早餐午餐晚餐其他"].values
typeOfMeal = df["中餐西餐"].values

# 提供給使用者的選擇
scoreChoice = [1,2,3,4,5]
timeChoice = [5,10,20,30]
priceChoice = [50,100,200,300,500]
mealChoice = ["早餐","午餐","晚餐","其他"]
typeOfMealChoice = ["中餐","西餐"]
locationChoice = ["北","中","南","東","全台"]

# 利用使用者的選擇從資料中挑選出合適的餐廳
def displayShop(三餐,中西餐,地區,價錢,等候時間,服務態度,食物品質):
    size = 0
    result = pd.DataFrame()
    for i in range(df.shape[0]):
        if(三餐 in meal[i] and 中西餐 in typeOfMeal[i] and 地區 in location[i]
           and 價錢 >= price[i] and 等候時間 >= waitTime[i] and 服務態度 <= service[i]
           and 食物品質 <= foodQuality[i]):
            result = result.append(numberRow(i+1))
            size+=1
    if(size==0):
        print("非常抱歉沒有符合您的選項的餐廳")
    else:
        print("\n有 " + str(size) + " 家餐廳符合你的要求ㄛ")
        # 另一種對字串的處理方式
        # print("\n有 {} 家餐廳符合你的要求ㄛ".format(size))      
        return result

# 處理互動表單的部分
interact(displayShop,三餐=mealChoice,中西餐=typeOfMealChoice,地區=locationChoice,價錢=priceChoice,等候時間=timeChoice,服務態度=scoreChoice,食物品質=scoreChoice)
```

```python
df["服務態度(1~5)"].plot(color = "green")
df["食物品質(1~5)"].plot(color = "yellow")
```

```python
analysisService=[]
for i in range(1,6):
    # 計算出不同數值的服務態度在總共的餐廳裡有幾個
    analysisService.append(df.loc[df["服務態度(1~5)"]==i].shape[0])
print(analysisService)
plt.bar(scoreChoice,analysisService)
```

```python
analysisFood=[]
for i in range(1,6):
    # 計算出不同數值的食物品質在總共的餐廳裡有幾個
    analysisFood.append(df.loc[df["食物品質(1~5)"]==i].shape[0])
print(analysisFood)
plt.bar(scoreChoice,analysisFood,color="brown")
```
