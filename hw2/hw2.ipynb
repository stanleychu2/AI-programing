```python
%matplotlib inline

import cv2
import numpy as np
import matplotlib.pyplot as plt
import imutils
```

```python
img = cv2.imread('dog.jpg')
image = imutils.resize(img, height=450)

plt.imshow(image)
```

```python
effect = [[[-1,-2,-1],[0, 0, 0],[1, 2, 1]], [[1,0,-1],[2,0,-2],[1, 0, -1]], [[1,2,1],[0, 0, 0],[-1, -2, -1]], [[-1,0,1],[-2,0,2],[-1, 0, 1]],
[[0,1,0],[1,-4,1],[0, 1, 0]], [[1, 1, 1],[1, -8, 1],[1, 1, 1]], [[1, 1, 1],[1, -7, 1],[1, 1, 1]], [[-1, -1, -1],[-1, 8, -1],[-1, -1, -1]], np.ones((3, 3), dtype=np.float32) / 3**2]

num = int(input("Kernel type: "))
```

```python
kernel = np.array(effect[num])
dst = cv2.filter2D(image, -1, kernel, anchor=(-1, -1), delta=0, borderType=cv2.BORDER_DEFAULT)

plt.imshow(dst)
```
