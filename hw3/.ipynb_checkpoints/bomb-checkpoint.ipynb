```python
%matplotlib inline

import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('./explode.png')
```

```python
def explode():
    rand = random.randint(2, 10)
    num = rand
    
    return num 
```

```python
def guess():
    player_ans = -1
    answer = random.randint(1, 100)
    count = explode()
    
    while(count!=0 and answer!=player_ans):
        count -= 1
        player_ans = int(input("請選擇一個數字(1~100)："))
        if(answer == player_ans):
            print("Win")
        elif(count==0):
            plt.imshow(img) 
```

```python
# 像電影中拆炸彈的場景一樣，你永遠不會知道現在剪掉這條
# 之後會爆炸還是不會，只能走一步算一步，有時候你就是不
# 會知道自己還剩多少時間還有多少機會
```

```python
guess()
```
